Triangle Classifier
Classify a triangle based on its side lengths.

Instructions
Read three side lengths from input (three lines).

First check if the sides can form a valid triangle. A triangle is valid if the sum of any two sides is greater than the third side.

If not valid, print Not a triangle.

If valid, print the type:

Equilateral if all three sides are equal
Isosceles if exactly two sides are equal
Scalene if no sides are equal

```python
# Read three sides
#a = int(input())
#b = int(input())
#c = int(input())
a = 3
b = 4
c = 5

# Classify and print
if a+b>c and b+c>a and a+c>b:
  if a==b==c:
    print("Equilateral")
  elif a==b!=c or b==c!=a or a==c!=b:
    print("Isosceles")
  elif a!=b!=c or b!=c!=a or a!=c!=b:
    print("Scalene")
else:
  print("Not a triangle")
```

Leap Year
Determine if a given year is a leap year.

Instructions
Read a year from input.

Print Leap year or Not a leap year.

A year is a leap year if:

It is divisible by 4 and
It is NOT divisible by 100, unless it is also divisible by 400

```python
#If a year is divisible by 4, it is a leap year... unless it ends in 00.
#If it ends in 00, it is only a leap year if it is also divisible by 400.

# Read the year
#year = int(input())
year = 2100

if (year%400)==0:
  print("Leap year")
elif (year%4)==0 and (year%100)!=0:
  print("Leap year")
else:
  print("Not a leap year")
```

Make a Username
Create a username and initials from a first and last name.

Instructions
Read a first name and a last name from input.

Create a username by joining the two names together in lowercase (no space between them).

Print these two lines:

Username: [username]
Initials: [first letter of first name][first letter of last name] (in uppercase)

```python
# Read input
first_name = input().strip()
last_name = input().strip()

# Create username (lowercase, no space)
username = (first_name+last_name).lower()
# Create initials (uppercase first letters)
initials = (first_name[0]+last_name[0]).upper()
# Print results
print(f"Username: {username}")
print(f"Initials: {initials}")
```

Factorial
Create a function that calculates the factorial of a number.

Instructions
Read a number from input.

Create a function that calculates the factorial of that number.

The factorial of N means: multiply all numbers from 1 to N together.

Example: factorial of 5 = 1 × 2 × 3 × 4 × 5 = 120

The factorial of 0 is 1.

Print the result like this:

[n]! = [result]


```python
# Create a function for factorial
def factorial(n):
    # Return the factorial of n
  result = 1
  for i in range(1, n+1):
    result = i * result
  return result

# Read input and print result
#num = int(input())
num = 5
print(str(num) + "! = " + str(factorial(num)))
```

Area Calculator
Calculate the area of a rectangle, triangle, or circle.

Instructions
Read a shape and its measurements from input.

The shape is one of these three words:

rectangle - next two lines are width and height
triangle - next two lines are base and height
circle - next line is the radius
Print the area with two decimal places:

Area: [result]
Triangle area = base × height / 2. Circle area = pi × radius × radius.

```python
import math

# Read the shape first
shape = input().strip()

# Read measurements and calculate area based on the shape
if shape == "rectangle":
    width = int(input())
    height = int(input())
    result = width * height

elif shape == "triangle":
    base = int(input())
    height = int(input())
    result = (base * height) / 2

elif shape == "circle":
    radius = int(input())
    result = math.pi * (radius ** 2)  # Radius squared!

else:
    print("wrong")

# Print the area (if it's a valid shape)
if shape in ["rectangle", "triangle", "circle"]:
    print(f"Area: {result:.2f}")
```

Rectangle Border
Print a rectangle border made of stars.

Instructions
Read a width and a height from input.

Print a rectangle border using * characters.

The first and last rows are full rows of stars. The rows in between have a star at the start and end, with spaces in the middle.

```python
# Read input
width = int(input())
height = int(input())

# Print the rectangle border
space = " " * (width -2)

for row in range(1, height+1):
  if row == 1 or row == height:
    print(width * "*")
  else:
    print("*"+space+"*")
```
